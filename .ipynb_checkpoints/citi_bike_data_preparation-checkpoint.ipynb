```python
import pandas as pd
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 2000)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('/Users/jmirabito/Desktop/Learn.co/mod_2_final_project/citi_bike_revenue_predictions/citi_bike_nj_clean.csv')
```

```python
print(df.shape)
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df.reset_index(inplace=True)
```

```python
df.rename(columns={'dist_mta': 'dist_PATH'}, inplace=True)
```

```python
print(df.shape)
df.head()
```

```python
# Removing unnecessary columns for linear regression
df.groupby(['date','station_id'])['rider'].sum()
```

```python
# Creating ridership column groubed by day and station
df_modeling = df.groupby(['date', 'station_id'])['rider'].sum()
```

```python
df_modeling = df_modeling.reset_index()
```

```python
df_modeling.set_index(['date','station_id'], inplace=True)
```

```python
df_modeling.rename(columns={'rider': 'rider_count'}, inplace=True)
```

```python
df_modeling.head()
```

```python
# Creating average temperature column grouped by station and date. Value will be same for all stations in a day.
df_modeling['TAVG'] = df.groupby(['date', 'station_id'])['TAVG'].median()
```

```python
#Creating daily precipitation column
df_modeling['PRCP'] = df.groupby(['date', 'station_id'])['PRCP'].median()
```

```python
df_modeling.PRCP.fillna(0.0, inplace=True)
```

```python
# Creating distance to PATH train column
df_modeling['dist_PATH'] = df.groupby(['date', 'station_id'])['dist_PATH'].median()
```

```python
# Creating distance to nearest landmark column
df_modeling['dist_landmark'] = df.groupby(['date', 'station_id'])['dist_landmark'].median()
```

```python
# Creating zipcode median income column
df_modeling['median_inc'] = df.groupby(['date', 'station_id'])['Median'].median()
```

```python
# Creating zipcode mean income column
df_modeling['mean_inc'] = df.groupby(['date', 'station_id'])['Mean'].median()
```

```python
# Creating zipcode population column
df_modeling['population_zip'] = df.groupby(['date', 'station_id'])['Pop'].median()
```

```python
# Creating zipcode column
df_modeling['station_zip'] = df.groupby(['date', 'station_id'])['citi_zipcode'].median()
```

```python
# Converting zipcodes to string format 
df_modeling.station_zip = df_modeling.station_zip.astype(str)
df_modeling.station_zip = df_modeling.station_zip.apply(lambda x: x[0:5])
df_modeling.station_zip = df_modeling.station_zip.apply(lambda x: x.strip('.'))
df_modeling.station_zip = df_modeling.station_zip.apply(lambda x: '0'+ x if x.startswith('7') else x)
```

```python
# Creating weekend dummy column
df_modeling['weekend'] = df.groupby(['date', 'station_id'])['weekend'].median()
```

```python
# Creating day of week dummies inoriginal df
df = pd.concat([df, pd.get_dummies(df['day'])], 1)
```

```python
# Now bringing them into modeling df
df_modeling['Monday'] = df.groupby(['date', 'station_id'])['Monday'].median()
df_modeling['Tuesday'] = df.groupby(['date', 'station_id'])['Tuesday'].median()
df_modeling['Wednesday'] = df.groupby(['date', 'station_id'])['Wednesday'].median()
df_modeling['Thursday'] = df.groupby(['date', 'station_id'])['Thursday'].median()
df_modeling['Friday'] = df.groupby(['date', 'station_id'])['Friday'].median()
df_modeling['Saturday'] = df.groupby(['date', 'station_id'])['Saturday'].median()
df_modeling['Sunday'] = df.groupby(['date', 'station_id'])['Sunday'].median()
```

```python
df_modeling.to_csv('citibike_modeling_dataset.csv')
```
