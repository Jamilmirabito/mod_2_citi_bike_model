```python
import pandas as pd
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr

pd.set_option("display.max_rows", 101)
pd.set_option("display.max_columns", 101)
```

```python
df = pd.read_csv('/Users/markdetiberiis/Desktop/Citibike_data/df_ready_for_eda.csv')
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
# Modfying zipcode to the correct format
df.citi_zipcode = df.citi_zipcode.astype(str)
df.citi_zipcode = df.citi_zipcode.apply(lambda x: x[0:5])
df.citi_zipcode = df.citi_zipcode.apply(lambda x: x.strip('.'))
df.citi_zipcode = df.citi_zipcode.apply(lambda x: '0'+ x if x.startswith('7') else x)
```

```python
df = df[~df.citi_zipcode.str.startswith("1")]
```

```python
df.corr(method = 'spearman')
```

```python
df_temp = df[['TAVG', 'rider']].copy()
```

```python
corr_temp = df.groupby(['TAVG']).corr(method = 'spearman', min_periods = 1)
corr_temp
```

AVG Daily Riders & Precipitation

```python
#avg daily riders by PRCP

df_rain = pd.DataFrame(df.groupby(['PRCP', 'date'])['rider'].sum().median(level='PRCP'))
```

```python
df_rain['prcp'] = df_rain.index
```

```python
#Pearson correlation prcp v avg daily riders

corr_prcp = df_rain.corr(method = 'pearson', min_periods = 1)
corr_prcp
```

AVG Daily Riders & TAVG

```python
#avg daily riders by TAVG

df_TAVG = pd.DataFrame(df.groupby(['TAVG', 'date'])['rider'].sum().median(level='TAVG'))
```

```python
df_TAVG['tavg'] = df_TAVG.index
```

```python
#Pearson correlation TAVG v avg daily riders

corr_TAVG = df_TAVG.corr(method = 'pearson', min_periods = 1)
corr_TAVG
```

Riders & Income

```python
df_income = pd.DataFrame(df.groupby(['Median', 'date'])['rider'].sum().median(level='Median')) 
```

```python
df_income['Median'] = df_income.index
```

```python
corr_income = df_income.corr(method = 'pearson', min_periods = 1)
corr_income
```
